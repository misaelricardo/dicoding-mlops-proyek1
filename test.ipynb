```python
import requests
import tensorflow as tf
import numpy as np
import json
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\bisto\tfx-dicoding\misaelricardo-dicoding-submission\data\winequality-red-comma.csv')
df
```

```python
data_to_predict = df.iloc[0, :-1]
data_to_predict
test_data = data_to_predict.to_dict()
```

```python
data = json.dumps({"instances": [test_data]})
print(data)
```

```python
response = requests.post("http://localhost:8080/v1/models/red-wine-quality-model:predict", data=data)
```

```python
if response.status_code == 200:
    predictions = response.json()["predictions"]
    # Assuming the prediction is a floating point number
    # We need to round it to the nearest integer
    predicted_quality = round(predictions[0][0])
    print("Predicted quality:", predicted_quality)
else:
    print("Error:", response.text)
```
